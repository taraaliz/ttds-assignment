### Plotting a correlation matrix

```python
# plotting a correlation matrix
spearman_corr = df.drop(columns=['rerun_ID']).corr(method='spearman')
print(spearman_corr)
ax = sns.heatmap(
    spearman_corr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(10, 250, n=200),
    square=True, annot=False, fmt='.1f', linewidths=.5
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
ax;
```

```python
fig = plt.figure()

# make space for a 5x3 grid of graphs, enough for 15 (we have 14 to plot)
fig, axs = plt.subplots(nrows=7, ncols=2, figsize=(16,32))

# unpack all the axes subplots
axs = axs.ravel()

# assign the plot to each subplot in axe
for i, c in enumerate(df.columns):
    X[c].plot(kind='box', title=c, ax=axs[i])
```

```python
for i in X.columns:
    plt.figure()
    plt.title(f'{i}')
    df[i].plot(kind='hist')
```
